## IOU metric, Initial data split and model 

Dictionary with each label

````python
name2id = {v:k for k,v in enumerate(codes)}; name2id
````

In this code we are going to use the Intersection Over Union (IOU) metric. A function to save predictions definitions is created aswell.

````python
def iou_metric(input, target):
    target = target.squeeze(1)
    mask = target != 0
    return (input.argmax(dim=1)[mask]==target[mask]).float().mean()

def save_preds(dl):
    i=0
    names = dl.dataset.items

    for b in dl:
      preds = learn.pred_batch(batch=b, reconstruct=True)
      for o in preds:
          o.save(path_gen/names[i].name)
          i += 1
````

Here we can set the weight decay, our metric, choose net model (resnet34 in this case) and choose if we want to use imageNet weights



````python
wd = 1e-2
metrics = iou_metric
used_model = models.resnet34
useImageNet = False
````

Define the type of data spliting (random with some porcentage is current being employed)

````python
src = (SegmentationItemList.from_folder(path_img)
       .split_by_rand_pct()
       .label_from_func(get_y_fn, classes=codes))
````

## Training and validation

Where we start the training part. First we employ the 375x250(1/4) resolution. The same exact sequence is performed for the 750x500(1/2) and 1500x1000(full) resolutions.

###375x250

Define the batch size and the resolution employed. A batch of size 2 has shown to be the best choice in order to avoid RAM overflow.

````python
size = src_size//4
size[1] = size[1]+1
bs = 2
````

Apply transformations such as resolution change and data augmentation to the data

````python
normalizePar=None

if useImageNet:
  normalizePar=imagenet_stats

  
data = (src.transform(get_transforms(), size=size, tfm_y=True)
        .databunch(bs=bs)
        .normalize(normalizePar))
````

Now we can see a batch.

````python
data.show_batch(2, figsize=(10,7))
````

Load the model with the data, the metric and the weight decay selected

````python
learn = unet_learner(data, used_model, metrics=metrics, wd=wd)
````

Find the most appropriate learning rate

````python
lr_find(learn)
learn.recorder.plot()
````

````python
lr=1e-3
````

Start the training with 5 cycles and the learning rate previously defined. It is not necessary to use too many cycles with the 1/4 resolution.

````python
learn.fit_one_cycle(5, slice(lr), pct_start=0.9)
````

Save the weitghs and load it to continue the trainning and perform some data release.

````python
learn.save('stage1-375x250')
````

````python
learn.load('stage1-375x250')
````

Unfreeze the learner to modify some early layers weight.

````python
learn.unfreeze()
````

  Use Differential Learning Rate (DLR) and continue training the learner. 10 more cycles must be enough.

````python
lrs = slice(lr/400,lr/4)
````

````python
learn.fit_one_cycle(10, lrs, pct_start=0.8)
````

Save the weitghs and load it to continue the trainning and perform some data release

````python
learn.save('stage2-375x250')
````

````python
del data
del learn
torch.cuda.empty_cache()
gc.collect()
````

# Post segmentation processing

Now we can use the trained model to predict some images and use classic computer vision algorithms to tune the results.

Here we have the necessary libraries.

````python
import cv2
import numpy as np
from matplotlib import pyplot as plt
````

The next fragment of code will get one image of our dataset, predict it and convert its format so we can work with it in OpenCV lib.

````python
img = data.train_ds[12][0]
img.show(y=learn.predict(img)[0],figsize=(12.45,16))
np_arr = learn.predict(img)[1].cpu().detach().numpy()
np_arr2 = np_arr[0]
np_arr2 = np_arr2.astype('uint8')
plt.figure(figsize = (12,10))
plt.imshow(np_arr2,'gray')
````

````python
type(img)
````

Next step is to apply closing and opening morphological operations to this images.

````python
kernel = np.ones((5,5),np.uint8)
np_arr2 = np_arr2.astype('uint8')
mask2 = cv2.morphologyEx(np_arr2, cv2.MORPH_CLOSE, kernel)
mask2 = cv2.morphologyEx(mask2, cv2.MORPH_OPEN, kernel)
kernel = np.ones((5,5),np.uint8)
np_arr2 = np_arr2.astype('uint8')
mask2 = cv2.morphologyEx(np_arr2, cv2.MORPH_CLOSE, kernel)
mask3 = cv2.morphologyEx(mask2, cv2.MORPH_OPEN, kernel)

plt.figure(figsize = (12,10))
plt.imshow(mask3,'gray')
````

A analysis of conected components is executed in order to eliminate elements who don't belong to the main wood log.

After the conected components analysis a convex hull algorithm is apllied. The result is a much smoother form, as indicated in the images.

````python
contours,hierarchy = cv2.findContours(mask3, cv2.CHAIN_APPROX_NONE, 2)
try:
    cnt = contours[0]
except:
    print("Nenhum objeto detectado na imagem!")
maior_area = 0
maior_area_indice = 0
for areas in range(len(contours)):
    N = cv2.contourArea(contours[areas])
    if N > maior_area:
        maior_area = N
        maior_area_indice = areas
hull = cv2.convexHull(contours[maior_area_indice])
cv2.fillPoly(mask3, pts = [hull], color=(100,100,100))
fig, axes = plt.subplots(1, 2, figsize=(50, 40), sharex=True, sharey=True)
ax = axes.ravel()
ax[0].imshow(np_arr2)
ax[1].imshow(mask3)
````
